```python
from haversine import inverse_haversine, Direction
```

```python
# Set point (Should be center of state)
bangalore = (13,78) # (lat, long)
```

```python
# GET THE LIST OF LOCATIONS

def get_list_of_loc(step):
    list_loc = []
    for step in range (0, 350, step):
        list_loc.append(inverse_haversine(bangalore, step, Direction.WEST))
    return list_loc
```

```python
print(get_list_of_loc(10))
```
